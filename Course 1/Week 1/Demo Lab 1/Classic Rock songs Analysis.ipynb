### Importing Libraries

```python
import pandas as pd
import sqlite3 as sq3
import pandas.io.sql as pds
print('done')
```

### Database Connection

```python
datapath = 'data/classic_rock.db'
con = sq3.Connection(datapath)
```

### Reading Data

```python
query = '''
SELECT *
FROM rock_songs;
'''
observations = pds.read_sql(query,con)
observations
```

### Counting the rows of data

```python
counting = '''
SELECT COUNT(*) FROM rock_songs; '''
counting_op = pds.read_sql(counting, con)
counting_op
```

```python
query = ''' SELECT Artist, Release_Year, COUNT(*) AS num_songs, AVG(PlayCount) AS avg_plays  
    FROM rock_songs
    GROUP BY Artist, Release_Year
    ORDER BY num_songs desc;
'''
observations_1 = pds.read_sql(query,con,
                           coerce_float = True,
                           parse_dates = ['Release_Year'],
                           chunksize = 5)
observations = pds.read_sql(query, con)

for index,observations in enumerate(observations_1):
    if index < 5:
        print(f'Observations index: {index}'.format(index))
        display(observations)
```
